```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Setup & Dataset creation

np.random.seed(0)
n = 200

df = pd.DataFrame({
    'age':      np.random.randint(18, 60, n),
    'income':   np.random.randint(20000, 120000, n),
    'spend':    np.random.randint(500, 8000, n),
    'category': np.random.choice(['Electronics', 'Clothing', 'Groceries', 'Sports'], n)
})

df.head()
```

# Task 1 — Bar Chart: Average Spend by Category

### Purpose: Compare average spending across categories

```python
avg_spend = df.groupby('category')['spend'].mean()

avg_spend.plot(kind='bar')
plt.title('Average Spend by Category')
plt.xlabel('Category')
plt.ylabel('Average Spend')
plt.show()
```

# Observation:
## The category with the highest average spend is Electronics.

# Task 2 — Scatter Plot: Income vs Spend

### Purpose: Identify relationship between income and spending

```python
sns.scatterplot(x=df['income'], y=df['spend'])
plt.title("Income vs Spend")
plt.xlabel("Income")
plt.ylabel("Spend")
plt.show()
```

# Observation:
## There is no clear strong relationship between income and spend, as the points are widely scattered.

# Task 3 — Interactive Scatter Plot with Plotly

### Purpose: Explore data interactively with categories

```python
import plotly.express as px

# Interactive (Colab ke liye)
fig = px.scatter(df, x='income', y='spend', color='category',
                 title="Interactive Income vs Spend")
fig.show()

```

```python
import matplotlib.pyplot as plt

# Static graph (GitHub ke liye)
for cat in df['category'].unique():
    subset = df[df['category'] == cat]
    plt.scatter(subset['income'], subset['spend'], label=cat)

plt.title("Income vs Spend (Static)")
plt.xlabel("Income")
plt.ylabel("Spend")
plt.legend()
plt.show()
```

#Observation: Interactive plot allows zooming and category filtering, while the static plot ensures visibility on GitHub.
